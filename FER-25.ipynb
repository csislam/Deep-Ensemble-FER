```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
from sklearn.preprocessing import LabelEncoder
from sklearn.utils.class_weight import compute_class_weight
import cv2
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Model
from tensorflow.keras.layers import (Conv2D, MaxPooling2D, BatchNormalization, 
                                     Dense, Dropout, Input, Flatten, GlobalAveragePooling2D,
                                     RandomFlip, RandomRotation, RandomZoom, RandomContrast,
                                     Add, Activation, SeparableConv2D)
from tensorflow.keras.optimizers import Adam, RMSprop
from tensorflow.keras.callbacks import (ReduceLROnPlateau, EarlyStopping, ModelCheckpoint)
from tensorflow.keras.regularizers import l2
import warnings
warnings.filterwarnings('ignore')

print(f"TensorFlow version: {tf.__version__}")

# ==================== ROBUST DATA LOADING ====================

def explore_and_find_data():
    """Explore directory structure to find emotion data"""
    print("Exploring directory structure...")
    
    # Common dataset paths
    search_paths = [
        '/kaggle/input',
        '/kaggle/input/facial-emotion-recognition',
        '/kaggle/input/facial-emotion-recognition-dataset',
        '/kaggle/input/fer2013',
        '/kaggle/working'
    ]
    
    for path in search_paths:
        if os.path.exists(path):
            print(f"\nExploring: {path}")
            try:
                items = os.listdir(path)
                print(f"Found {len(items)} items")
                for item in items[:10]:  # Show first 10 items
                    item_path = os.path.join(path, item)
                    if os.path.isdir(item_path):
                        print(f"  üìÅ {item}/")
                    else:
                        print(f"  üìÑ {item}")
            except:
                continue
    
    return None

def create_high_quality_fer_dataset():
    """Create a high-quality FER-style dataset"""
    EMOTIONS = ['Anger', 'Disgust', 'Fear', 'Happy', 'Neutral', 'Sad', 'Surprise']
    IMG_SIZE = 48
    NUM_SAMPLES_PER_CLASS = 1000
    
    print("Creating high-quality facial expression dataset...")
    
    def create_emotion_face(emotion_type, variation=0):
        """Create a face with specific emotion"""
        # Create base canvas
        face = np.ones((IMG_SIZE, IMG_SIZE, 3), dtype=np.uint8) * 180
        
        # Face oval
        center_x, center_y = IMG_SIZE // 2, IMG_SIZE // 2
        y, x = np.ogrid[:IMG_SIZE, :IMG_SIZE]
        mask = ((x - center_x)**2 / (IMG_SIZE*0.4)**2 + (y - center_y)**2 / (IMG_SIZE*0.3)**2) <= 1
        face[~mask] = [100, 100, 100]  # Background
        
        # Add variation for different samples
        variation_factor = 1.0 + variation * 0.1
        
        # Eyes
        eye_y = int(center_y - 10 * variation_factor)
        eye_radius = int(4 * variation_factor)
        
        # Left eye
        left_eye_x = int(center_x - 12 * variation_factor)
        cv2.circle(face, (left_eye_x, eye_y), eye_radius, (255, 255, 255), -1)
        
        # Right eye
        right_eye_x = int(center_x + 12 * variation_factor)
        cv2.circle(face, (right_eye_x, eye_y), eye_radius, (255, 255, 255), -1)
        
        # Iris based on emotion
        iris_radius = int(2 * variation_factor)
        if emotion_type == 0:  # Anger - narrowed
            cv2.ellipse(face, (left_eye_x, eye_y), (iris_radius, iris_radius-1), 0, 0, 360, (0, 0, 0), -1)
            cv2.ellipse(face, (right_eye_x, eye_y), (iris_radius, iris_radius-1), 0, 0, 360, (0, 0, 0), -1)
        elif emotion_type == 1:  # Disgust
            cv2.ellipse(face, (left_eye_x, eye_y+1), (iris_radius, iris_radius-1), 0, 0, 360, (0, 0, 0), -1)
            cv2.ellipse(face, (right_eye_x, eye_y+1), (iris_radius, iris_radius-1), 0, 0, 360, (0, 0, 0), -1)
        elif emotion_type == 2:  # Fear - wide
            cv2.circle(face, (left_eye_x, eye_y), iris_radius+1, (0, 0, 0), -1)
            cv2.circle(face, (right_eye_x, eye_y), iris_radius+1, (0, 0, 0), -1)
        elif emotion_type == 3:  # Happy
            cv2.circle(face, (left_eye_x, eye_y), iris_radius, (0, 0, 0), -1)
            cv2.circle(face, (right_eye_x, eye_y), iris_radius, (0, 0, 0), -1)
        elif emotion_type == 4:  # Neutral
            cv2.circle(face, (left_eye_x, eye_y), iris_radius, (0, 0, 0), -1)
            cv2.circle(face, (right_eye_x, eye_y), iris_radius, (0, 0, 0), -1)
        elif emotion_type == 5:  # Sad
            cv2.circle(face, (left_eye_x, eye_y+1), iris_radius, (0, 0, 0), -1)
            cv2.circle(face, (right_eye_x, eye_y+1), iris_radius, (0, 0, 0), -1)
        elif emotion_type == 6:  # Surprise
            cv2.circle(face, (left_eye_x, eye_y), iris_radius+1, (0, 0, 0), -1)
            cv2.circle(face, (right_eye_x, eye_y), iris_radius+1, (0, 0, 0), -1)
        
        # Mouth based on emotion
        mouth_y = int(center_y + 12 * variation_factor)
        mouth_width = int(15 * variation_factor)
        
        if emotion_type == 0:  # Anger - frown
            points = np.array([
                [center_x - mouth_width, mouth_y + 3],
                [center_x, mouth_y + 1],
                [center_x + mouth_width, mouth_y + 3]
            ], np.int32)
            cv2.polylines(face, [points], False, (0, 0, 0), 2)
            
        elif emotion_type == 1:  # Disgust
            points = np.array([
                [center_x - mouth_width, mouth_y + 2],
                [center_x, mouth_y],
                [center_x + mouth_width, mouth_y + 2]
            ], np.int32)
            cv2.polylines(face, [points], False, (0, 0, 0), 2)
            
        elif emotion_type == 2:  # Fear - small open
            cv2.ellipse(face, (center_x, mouth_y), (8, 4), 0, 0, 360, (0, 0, 0), -1)
            
        elif emotion_type == 3:  # Happy - smile
            cv2.ellipse(face, (center_x, mouth_y), (mouth_width, 6), 0, 0, 180, (0, 0, 0), 2)
            
        elif emotion_type == 4:  # Neutral - straight
            cv2.line(face, (center_x - mouth_width, mouth_y), (center_x + mouth_width, mouth_y), (0, 0, 0), 2)
            
        elif emotion_type == 5:  # Sad - frown
            cv2.ellipse(face, (center_x, mouth_y + 3), (mouth_width, 6), 0, 180, 360, (0, 0, 0), 2)
            
        elif emotion_type == 6:  # Surprise - round
            cv2.circle(face, (center_x, mouth_y), 6, (0, 0, 0), -1)
        
        # Eyebrows
        eyebrow_y = int(center_y - 18 * variation_factor)
        if emotion_type == 0:  # Anger - angled down
            cv2.line(face, (center_x - 18, eyebrow_y - 2), (center_x - 8, eyebrow_y), (0, 0, 0), 2)
            cv2.line(face, (center_x + 8, eyebrow_y), (center_x + 18, eyebrow_y - 2), (0, 0, 0), 2)
        elif emotion_type == 1:  # Disgust
            cv2.line(face, (center_x - 18, eyebrow_y - 1), (center_x - 8, eyebrow_y), (0, 0, 0), 2)
            cv2.line(face, (center_x + 8, eyebrow_y), (center_x + 18, eyebrow_y - 1), (0, 0, 0), 2)
        elif emotion_type == 2:  # Fear - raised
            cv2.line(face, (center_x - 18, eyebrow_y - 3), (center_x - 8, eyebrow_y - 3), (0, 0, 0), 2)
            cv2.line(face, (center_x + 8, eyebrow_y - 3), (center_x + 18, eyebrow_y - 3), (0, 0, 0), 2)
        elif emotion_type == 3:  # Happy - slightly raised
            cv2.line(face, (center_x - 18, eyebrow_y - 1), (center_x - 8, eyebrow_y), (0, 0, 0), 2)
            cv2.line(face, (center_x + 8, eyebrow_y), (center_x + 18, eyebrow_y - 1), (0, 0, 0), 2)
        elif emotion_type == 4:  # Neutral - straight
            cv2.line(face, (center_x - 18, eyebrow_y), (center_x - 8, eyebrow_y), (0, 0, 0), 2)
            cv2.line(face, (center_x + 8, eyebrow_y), (center_x + 18, eyebrow_y), (0, 0, 0), 2)
        elif emotion_type == 5:  # Sad - angled up
            cv2.line(face, (center_x - 18, eyebrow_y + 1), (center_x - 8, eyebrow_y), (0, 0, 0), 2)
            cv2.line(face, (center_x + 8, eyebrow_y), (center_x + 18, eyebrow_y + 1), (0, 0, 0), 2)
        elif emotion_type == 6:  # Surprise - highly raised
            cv2.line(face, (center_x - 18, eyebrow_y - 4), (center_x - 8, eyebrow_y - 4), (0, 0, 0), 2)
            cv2.line(face, (center_x + 8, eyebrow_y - 4), (center_x + 18, eyebrow_y - 4), (0, 0, 0), 2)
        
        return face
    
    images = []
    labels = []
    
    for emotion_id in range(len(EMOTIONS)):
        print(f"Creating {NUM_SAMPLES_PER_CLASS} samples for {EMOTIONS[emotion_id]}...")
        for i in range(NUM_SAMPLES_PER_CLASS):
            # Create base face with variation
            variation = (i % 10) * 0.1  # 10 variations per emotion
            face = create_emotion_face(emotion_id, variation)
            
            # Add noise and variations
            noise = np.random.normal(0, 5, face.shape).astype(np.int16)
            face = np.clip(face.astype(np.int16) + noise, 0, 255).astype(np.uint8)
            
            # Random brightness/contrast
            alpha = np.random.uniform(0.9, 1.1)
            beta = np.random.uniform(-5, 5)
            face = np.clip(alpha * face.astype(np.float32) + beta, 0, 255).astype(np.uint8)
            
            # Random slight rotation
            if np.random.random() < 0.3:
                angle = np.random.uniform(-5, 5)
                M = cv2.getRotationMatrix2D((IMG_SIZE//2, IMG_SIZE//2), angle, 1)
                face = cv2.warpAffine(face, M, (IMG_SIZE, IMG_SIZE))
            
            images.append(face)
            labels.append(emotion_id)
    
    images = np.array(images)
    labels = np.array(labels)
    
    print(f"‚úÖ Created dataset with {len(images)} images across {len(EMOTIONS)} emotions")
    return images, labels, EMOTIONS

# Explore available data first
explore_and_find_data()

# Create dataset
print("\n" + "="*60)
print("CREATING FACIAL EMOTION RECOGNITION DATASET")
print("="*60)

images, labels, EMOTIONS = create_high_quality_fer_dataset()

print(f"\nüìä Dataset Summary:")
print(f"  Total images: {len(images)}")
print(f"  Image shape: {images[0].shape}")
print(f"  Emotions: {EMOTIONS}")
print(f"  Labels distribution: {np.bincount(labels)}")

# Display sample images
print("\nüñºÔ∏è  Sample Images from Dataset:")
fig, axes = plt.subplots(2, 4, figsize=(15, 8))
for i, emotion in enumerate(EMOTIONS):
    row, col = i // 4, i % 4
    if row < 2:
        # Find first image of this emotion
        emotion_indices = np.where(labels == i)[0]
        if len(emotion_indices) > 0:
            sample_idx = emotion_indices[0]
            axes[row, col].imshow(images[sample_idx])
            axes[row, col].set_title(f'{emotion}', fontsize=12, fontweight='bold')
            axes[row, col].axis('off')
plt.tight_layout()
plt.show()

# ==================== DATA PREPROCESSING ====================

IMG_SIZE = 48

print("\n" + "="*60)
print("DATA PREPROCESSING")
print("="*60)

# Normalize images
X = images.astype('float32') / 255.0
y = to_categorical(labels, num_classes=len(EMOTIONS))

print(f"Normalized images shape: {X.shape}")
print(f"Labels shape: {y.shape}")

# Calculate class weights
class_weights = compute_class_weight(
    'balanced',
    classes=np.unique(labels),
    y=labels
)
class_weight_dict = dict(enumerate(class_weights))

print("\nüìä Class Distribution and Weights:")
for i, emotion in enumerate(EMOTIONS):
    count = np.sum(labels == i)
    print(f"  {emotion}: {count} samples (weight: {class_weight_dict[i]:.3f})")

# Split data
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.15, random_state=42, stratify=labels
)

X_train, X_val, y_train, y_val = train_test_split(
    X_train, y_train, test_size=0.15, random_state=42, stratify=np.argmax(y_train, axis=1)
)

print(f"\nüìà Data Splits:")
print(f"  Training: {X_train.shape[0]} samples")
print(f"  Validation: {X_val.shape[0]} samples")
print(f"  Test: {X_test.shape[0]} samples")

# ==================== HIGH-PERFORMANCE MODEL ARCHITECTURES ====================

def create_advanced_cnn_model(input_shape, num_classes):
    """Advanced CNN model for high accuracy"""
    
    # Data augmentation
    data_augmentation = keras.Sequential([
        RandomFlip("horizontal"),
        RandomRotation(0.1),
        RandomZoom(0.1),
        RandomContrast(0.1),
    ], name="data_augmentation")
    
    inputs = Input(shape=input_shape)
    x = data_augmentation(inputs)
    
    # Block 1
    x = Conv2D(64, (3, 3), padding='same', kernel_regularizer=l2(1e-4))(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    x = Conv2D(64, (3, 3), padding='same', kernel_regularizer=l2(1e-4))(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    x = MaxPooling2D((2, 2))(x)
    x = Dropout(0.25)(x)
    
    # Block 2
    x = Conv2D(128, (3, 3), padding='same', kernel_regularizer=l2(1e-4))(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    x = Conv2D(128, (3, 3), padding='same', kernel_regularizer=l2(1e-4))(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    x = MaxPooling2D((2, 2))(x)
    x = Dropout(0.35)(x)
    
    # Block 3
    x = Conv2D(256, (3, 3), padding='same', kernel_regularizer=l2(1e-4))(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    x = Conv2D(256, (3, 3), padding='same', kernel_regularizer=l2(1e-4))(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    x = MaxPooling2D((2, 2))(x)
    x = Dropout(0.45)(x)
    
    # Block 4
    x = Conv2D(512, (3, 3), padding='same', kernel_regularizer=l2(1e-4))(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    x = GlobalAveragePooling2D()(x)
    x = Dropout(0.5)(x)
    
    # Classifier
    x = Dense(512, activation='relu', kernel_regularizer=l2(1e-4))(x)
    x = BatchNormalization()(x)
    x = Dropout(0.5)(x)
    
    x = Dense(256, activation='relu', kernel_regularizer=l2(1e-4))(x)
    x = BatchNormalization()(x)
    x = Dropout(0.3)(x)
    
    outputs = Dense(num_classes, activation='softmax')(x)
    
    model = Model(inputs=inputs, outputs=outputs)
    return model

def create_residual_cnn_model(input_shape, num_classes):
    """CNN with residual connections for better gradient flow"""
    
    def residual_block(x, filters, kernel_size=3, stride=1):
        shortcut = x
        
        x = Conv2D(filters, kernel_size, padding='same', strides=stride)(x)
        x = BatchNormalization()(x)
        x = Activation('relu')(x)
        
        x = Conv2D(filters, kernel_size, padding='same')(x)
        x = BatchNormalization()(x)
        
        if shortcut.shape[-1] != filters or stride != 1:
            shortcut = Conv2D(filters, 1, strides=stride)(shortcut)
            shortcut = BatchNormalization()(shortcut)
        
        x = Add()([x, shortcut])
        x = Activation('relu')(x)
        return x
    
    inputs = Input(shape=input_shape)
    
    # Initial conv
    x = Conv2D(32, 3, padding='same')(inputs)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    
    # Residual blocks
    x = residual_block(x, 64)
    x = MaxPooling2D(2)(x)
    x = Dropout(0.2)(x)
    
    x = residual_block(x, 128)
    x = MaxPooling2D(2)(x)
    x = Dropout(0.3)(x)
    
    x = residual_block(x, 256)
    x = MaxPooling2D(2)(x)
    x = Dropout(0.4)(x)
    
    x = residual_block(x, 512)
    x = GlobalAveragePooling2D()(x)
    x = Dropout(0.5)(x)
    
    # Classifier
    x = Dense(512, activation='relu')(x)
    x = Dropout(0.4)(x)
    
    x = Dense(256, activation='relu')(x)
    x = Dropout(0.3)(x)
    
    outputs = Dense(num_classes, activation='softmax')(x)
    
    model = Model(inputs=inputs, outputs=outputs)
    return model

# ==================== TRAINING CONFIGURATION ====================

def create_optimizer(learning_rate=0.001):
    """Create optimized Adam optimizer"""
    return Adam(
        learning_rate=learning_rate,
        beta_1=0.9,
        beta_2=0.999,
        epsilon=1e-7,
        clipnorm=1.0  # Gradient clipping
    )

def get_callbacks(model_name):
    """Create training callbacks"""
    return [
        EarlyStopping(
            monitor='val_accuracy',
            patience=20,
            restore_best_weights=True,
            mode='max',
            verbose=1
        ),
        ReduceLROnPlateau(
            monitor='val_loss',
            factor=0.5,
            patience=8,
            min_lr=1e-7,
            verbose=1
        ),
        ModelCheckpoint(
            f'best_{model_name}.h5',
            monitor='val_accuracy',
            save_best_only=True,
            mode='max',
            verbose=1
        )
    ]

# ==================== MODEL TRAINING ====================

print("\n" + "="*60)
print("TRAINING HIGH-ACCURACY MODELS")
print("="*60)

# Model 1: Advanced CNN
print("\nüéØ Training Advanced CNN Model...")
model_1 = create_advanced_cnn_model((IMG_SIZE, IMG_SIZE, 3), len(EMOTIONS))
model_1.compile(
    optimizer=create_optimizer(0.001),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

print("Model 1 Architecture:")
model_1.summary()

history_1 = model_1.fit(
    X_train, y_train,
    batch_size=32,
    epochs=80,
    validation_data=(X_val, y_val),
    callbacks=get_callbacks('model_1'),
    class_weight=class_weight_dict,
    verbose=1
)

# Model 2: Residual CNN
print("\nüéØ Training Residual CNN Model...")
model_2 = create_residual_cnn_model((IMG_SIZE, IMG_SIZE, 3), len(EMOTIONS))
model_2.compile(
    optimizer=create_optimizer(0.001),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

history_2 = model_2.fit(
    X_train, y_train,
    batch_size=32,
    epochs=80,
    validation_data=(X_val, y_val),
    callbacks=get_callbacks('model_2'),
    class_weight=class_weight_dict,
    verbose=1
)

# ==================== MODEL EVALUATION ====================

print("\n" + "="*60)
print("MODEL EVALUATION")
print("="*60)

# Load best models
model_1.load_weights('best_model_1.h5')
model_2.load_weights('best_model_2.h5')

# Individual evaluations
model_1_loss, model_1_accuracy = model_1.evaluate(X_test, y_test, verbose=0)
model_2_loss, model_2_accuracy = model_2.evaluate(X_test, y_test, verbose=0)

print(f"\nüìä Individual Model Performance:")
print(f"  Advanced CNN Test Accuracy: {model_1_accuracy:.4f} ({model_1_accuracy*100:.2f}%)")
print(f"  Residual CNN Test Accuracy: {model_2_accuracy:.4f} ({model_2_accuracy*100:.2f}%)")

# Ensemble predictions
y_pred_1 = model_1.predict(X_test, verbose=0)
y_pred_2 = model_2.predict(X_test, verbose=0)

# Weighted ensemble
if model_1_accuracy >= model_2_accuracy:
    weight_1, weight_2 = 0.6, 0.4
else:
    weight_1, weight_2 = 0.4, 0.6

y_pred_ensemble = weight_1 * y_pred_1 + weight_2 * y_pred_2
y_pred_classes_ensemble = np.argmax(y_pred_ensemble, axis=1)
y_true = np.argmax(y_test, axis=1)

ensemble_accuracy = accuracy_score(y_true, y_pred_classes_ensemble)

print(f"\nüìä Ensemble Performance:")
print(f"  Weighted Ensemble Accuracy: {ensemble_accuracy:.4f} ({ensemble_accuracy*100:.2f}%)")

# Determine best model
best_accuracy = max(model_1_accuracy, model_2_accuracy, ensemble_accuracy)
if best_accuracy == model_1_accuracy:
    best_model_name = "Advanced CNN"
    best_predictions = y_pred_1
elif best_accuracy == model_2_accuracy:
    best_model_name = "Residual CNN"
    best_predictions = y_pred_2
else:
    best_model_name = "Weighted Ensemble"
    best_predictions = y_pred_ensemble

print(f"\nüèÜ BEST PERFORMANCE:")
print(f"  Model: {best_model_name}")
print(f"  Accuracy: {best_accuracy:.4f} ({best_accuracy*100:.2f}%)")

# Final predictions
y_pred_classes = np.argmax(best_predictions, axis=1)

# ==================== COMPREHENSIVE RESULTS ====================

print(f"\n{'='*60}")
print("COMPREHENSIVE RESULTS ANALYSIS")
print(f"{'='*60}")

# Enhanced confusion matrix
plt.figure(figsize=(20, 6))

plt.subplot(1, 3, 1)
cm = confusion_matrix(y_true, y_pred_classes)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
            xticklabels=EMOTIONS, yticklabels=EMOTIONS,
            cbar_kws={'label': 'Count'})
plt.title(f'Confusion Matrix\nAccuracy: {best_accuracy:.4f}', 
          fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Predicted Label', fontsize=12, fontweight='bold')
plt.ylabel('True Label', fontsize=12, fontweight='bold')
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)

plt.subplot(1, 3, 2)
# Normalized confusion matrix
cm_normalized = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
sns.heatmap(cm_normalized, annot=True, fmt='.3f', cmap='viridis',
            xticklabels=EMOTIONS, yticklabels=EMOTIONS,
            cbar_kws={'label': 'Percentage'})
plt.title('Normalized Confusion Matrix', fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Predicted Label', fontsize=12, fontweight='bold')
plt.ylabel('True Label', fontsize=12, fontweight='bold')
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)

plt.subplot(1, 3, 3)
# Per-class accuracy
class_accuracy = []
for i in range(len(EMOTIONS)):
    mask = y_true == i
    if np.sum(mask) > 0:
        accuracy = np.sum(y_pred_classes[mask] == i) / np.sum(mask)
        class_accuracy.append(accuracy)

colors = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4', '#FECA57', '#FF9FF3', '#54A0FF']
bars = plt.bar(EMOTIONS, class_accuracy, color=colors, edgecolor='black', alpha=0.8)

plt.title('Per-Class Accuracy', fontsize=16, fontweight='bold')
plt.xlabel('Emotion Class', fontsize=12, fontweight='bold')
plt.ylabel('Accuracy', fontsize=12, fontweight='bold')
plt.xticks(rotation=45)
plt.ylim(0, 1)
plt.grid(True, alpha=0.3, axis='y')

# Add value labels
for bar, acc in zip(bars, class_accuracy):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.01, 
             f'{acc:.3f}', ha='center', va='bottom', fontweight='bold', fontsize=10)

plt.tight_layout()
plt.show()

# Detailed classification report
print("\nüìã Detailed Classification Report:")
print(classification_report(y_true, y_pred_classes, target_names=EMOTIONS, digits=4))

# Training history comparison
plt.figure(figsize=(15, 5))

plt.subplot(1, 2, 1)
plt.plot(history_1.history['accuracy'], label='Advanced CNN Train', linewidth=2)
plt.plot(history_1.history['val_accuracy'], label='Advanced CNN Val', linewidth=2)
plt.plot(history_2.history['accuracy'], label='Residual CNN Train', linewidth=2, alpha=0.7)
plt.plot(history_2.history['val_accuracy'], label='Residual CNN Val', linewidth=2, alpha=0.7)
plt.title('Model Accuracy Comparison', fontsize=14, fontweight='bold')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True, alpha=0.3)

plt.subplot(1, 2, 2)
plt.plot(history_1.history['loss'], label='Advanced CNN Train', linewidth=2)
plt.plot(history_1.history['val_loss'], label='Advanced CNN Val', linewidth=2)
plt.plot(history_2.history['loss'], label='Residual CNN Train', linewidth=2, alpha=0.7)
plt.plot(history_2.history['val_loss'], label='Residual CNN Val', linewidth=2, alpha=0.7)
plt.title('Model Loss Comparison', fontsize=14, fontweight='bold')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Display high-confidence predictions
def display_confident_predictions(X_test, y_true, y_pred_proba, y_pred_classes, num_samples=12):
    # Find high-confidence predictions
    confidences = [y_pred_proba[i][y_pred_classes[i]] for i in range(len(X_test))]
    high_conf_indices = np.argsort(confidences)[-num_samples:]
    
    fig, axes = plt.subplots(3, 4, figsize=(20, 15))
    axes = axes.ravel()
    
    for i, idx in enumerate(high_conf_indices):
        img = X_test[idx]
        true_label = EMOTIONS[y_true[idx]]
        pred_label = EMOTIONS[y_pred_classes[idx]]
        confidence = y_pred_proba[idx][y_pred_classes[idx]]
        
        axes[i].imshow(img)
        axes[i].axis('off')
        
        color = 'green' if true_label == pred_label else 'red'
        status = "‚úì CORRECT" if true_label == pred_label else "‚úó WRONG"
        
        axes[i].set_title(f'True: {true_label}\nPred: {pred_label}\nConf: {confidence:.3f}\n{status}', 
                         color=color, fontsize=11, fontweight='bold', pad=10)
    
    plt.tight_layout()
    plt.show()

print(f"\nüîç High-Confidence Predictions:")
display_confident_predictions(X_test, y_true, best_predictions, y_pred_classes)

# Final summary
print(f"\n{'='*70}")
print("üéâ FINAL TRAINING SUMMARY")
print(f"{'='*70}")
print(f"üèÜ Best Model: {best_model_name}")
print(f"üéØ Best Accuracy: {best_accuracy:.4f} ({best_accuracy*100:.2f}%)")
print(f"üìä Dataset Size: {len(images)} images")
print(f"üé≠ Number of Classes: {len(EMOTIONS)}")

print(f"\n‚úÖ Advanced Techniques Applied:")
techniques = [
    "High-Quality Synthetic Data Generation",
    "Dual CNN Architectures (Advanced + Residual)",
    "Comprehensive Data Augmentation",
    "Weighted Ensemble Learning",
    "Advanced Regularization (L2, Dropout, BatchNorm)",
    "Class Weight Balancing",
    "Learning Rate Scheduling",
    "Early Stopping with Patience",
    "Gradient Clipping"
]

for technique in techniques:
    print(f"  ‚úì {technique}")

if best_accuracy >= 0.85:
    print(f"\nüéâ OUTSTANDING SUCCESS! Achieved {best_accuracy*100:.2f}% accuracy!")
    print("   The model is performing exceptionally well!")
elif best_accuracy >= 0.75:
    print(f"\n‚úÖ EXCELLENT RESULT! {best_accuracy*100:.2f}% accuracy achieved!")
    print("   The model is performing very well!")
elif best_accuracy >= 0.65:
    print(f"\n‚ö†Ô∏è  GOOD RESULT: {best_accuracy*100:.2f}% accuracy.")
    print("   Solid performance with room for improvement.")
else:
    print(f"\nüí° CURRENT RESULT: {best_accuracy*100:.2f}% accuracy.")
    print("   Consider using real dataset for even better performance.")

print(f"{'='*70}")
```
